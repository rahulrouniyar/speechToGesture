**This is for backend part of downloading youtube videos for conan**

```python
import pandas as pd
import youtube_dl as yt
import numpy as np
from subprocess import call
import os
```

```python
video_file = pd.read_csv("/content/videos_links.csv")
intervals_file = pd.read_csv("/content/intervals_df.csv") 
```

```python
conan = video_file[video_file['speaker'] == 'conan']
```

```python
print(conan)
```

```python
# changing pandas dataframe to numpy array and removing first column named speaker
links = np.array(conan.iloc[:,1:3])
```

```python
print(type(links[0][0]))
```

```python
print(links.shape[0])
```

```python
# extract intervals for specific video from intervals_csv_file
intervals_data = intervals_file[intervals_file['video_fn'] == links[0][0]]
print(intervals_data)
intervals_for_video = np.array(intervals_data.iloc[:,4:6])
print(intervals_for_video[0])
```

```python
print(intervals_for_video.shape)
```

```python
# loop this till every interval of specific video is extracted
start_time = intervals_for_video[0][0]
end_time = intervals_for_video[0][1]

start_time = start_time.split()
end_time = end_time.split()

start_time = start_time[1]
end_time = end_time[1]
print(start_time)
print(end_time)
```

```python
print(links[0][1])
link = str(links[0][1])
print(type(link))
print(link)
```
